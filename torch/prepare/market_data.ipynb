```python
from sqlalchemy import create_engine
import pandas as pd
# Define the query
from datetime import datetime, timedelta

# Create the connection string
connection_string = f'mysql+pymysql://root:rootpassword@127.0.0.1:3301/catcher'

# Create the SQLAlchemy engine
engine = create_engine(connection_string)
```

```python

symbol = 'NEAR'
```

```python

# Define the query to get data for the last 24 hours
query = f"SELECT * FROM bybit_market_data where symbol = '{symbol}' order by created_at desc"

# Execute the query and convert the data to a pandas DataFrame
df = pd.read_sql(query, engine)

# Display the DataFrame
#print(df)
```

```python
# Get max value of lastPcnt
#max_lastPcnt = df['lastPcnt'].mean()

print(f"Mean lastPcnt: {df['lastPcnt'].mean()}")
print(f"Max lastPcnt: {df['lastPcnt'].max()}")
print(f"Min lastPcnt: {df['lastPcnt'].min()}")
```

```python
# Display the resampled data
import matplotlib.pyplot as plt

df['lastPcnt'].hist(bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of lastPcnt')
plt.show()
```
